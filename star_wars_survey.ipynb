# Analyzing Star Wars Survey Data

dataset: https://github.com/fivethirtyeight/data/tree/master/star-wars-survey

```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
star_wars.head(10)
```

```python
star_wars.columns
```

```python
# keeping only rows with valid Respondent IDs
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
```

```python
# converting yes / no values to booleans for ease of comparisons

movies_seen = 'Have you seen any of the 6 films in the Star Wars franchise?'
movies_fan = 'Do you consider yourself to be a fan of the Star Wars film franchise?'

bool_map = { 'Yes' : True, 'No' : False }

for col in [movies_seen, movies_fan]:
    star_wars[col] = star_wars[col].map(bool_map)
```

```python
star_wars.head()
```

```python
# mapping movie names to booleans
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

#star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
# renaming the columns to have more intuitive names
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

# star_wars.head()

```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

```python
#converted the ratings so that higher value represents better rating for more intuitive plotting
plt.bar(range(6), 6-star_wars[star_wars.columns[9:15]].mean())
```

# Observations
The original movies were rated higher with Episode V being the highest rated of the franchise. So people do agree that 'The Empire Strikes Back' is the best of the bunch..

```python
# Looking for the most seen movie
for col in ['seen_1', 'seen_2', 'seen_3', 'seen_4', 'seen_5', 'seen_6']:
    print(star_wars[col].sum())

plt.bar(range(6),star_wars[star_wars.columns[3:9]].sum())
```

So Episode V is the most viewed movie in the franchise. The original movies seem be seen more than the newer ones, which correlates to the older ones being ranked higher.

```python
# Splitting the data into two groups based on gender

males = star_wars[star_wars['Gender'] == 'Male'] 
females = star_wars[star_wars['Gender'] == 'Female']
```

```python
# finding the most viewed movies for each gender
plt.bar(range(6), males[star_wars.columns[3:9]].sum())
plt.show()
plt.bar(range(6), females[star_wars.columns[3:9]].sum())
plt.show()
```

```python
# Finding the highest ranked movies for each gender (6 being the highest rank)
plt.bar(range(6), 6-males[star_wars.columns[9:15]].mean())
plt.xlabel('movies')
plt.ylabel('rating')
plt.show()

plt.bar(range(6), 6-females[star_wars.columns[9:15]].mean())
plt.xlabel('movies')
plt.ylabel('rating')
plt.show()
```

Even though more males than females watched the first three movies, they rated these films signifcantly lower than females.
